## Purpose
The Bronze layer stores raw, immutable source data ingested from external
systems. Data is landed with minimal transformation to preserve full
fidelity and enable reproducibility.

## Design Principles
- Append-only ingestion
- No business logic or aggregations
- Schema captured as received from source
- Raw values preserved for traceability
- Ingestion metadata added

## Core Tables

### Bronze_Flight_Market
- Grain: One row per airline market ticket (as delivered by source)
- Source: External airline market datasets
- Columns:
  - itinerary and market identifiers
  - origin and destination attributes
  - carrier identifiers
  - fares and distance metrics
  - ingestion_ts
  - source_year
  - source_quarter

## Transformations Applied
- None (except technical ingestion metadata)
- File format normalization
- Basic schema capture

## Operational Notes
- Bronze data is never updated or deleted
- All corrections are handled in Silver or Gold
- Serves as the audit and replay layer for the pipeline
- Partitioned by source_year and source_quarter
