## Purpose
The Silver layer contains cleaned, standardized, and conformed datasets
derived from Bronze. Data quality rules are applied, schemas are stabilized,
and business keys are prepared for dimensional modeling.

## Design Principles
- One-to-one or reduced-grain transformation from Bronze
- No aggregations
- Deterministic, repeatable transformations
- Schema enforcement and type standardization
- Data quality checks applied before promotion to Gold

## Core Tables

### Silver_Flight_Market
- Grain: One row per airline market ticket
- Source: Bronze flight market data
- Key Columns:
  - itin_id
  - mkt_id
  - carrier codes
  - origin and destination airport identifiers
- Measures:
  - passenger_cnt
  - market_fare_usd
  - miles and distance attributes
- Partitions:
  - source_year
  - source_quarter

## Transformations Applied
- Column renaming and standardization
- Data type casting
- Duplicate removal
- Null handling and basic validations
- Derivation of year and quarter attributes

## Operational Notes
- Silver is the single source for all Gold facts and dimensions
- No business aggregations exist in this layer
- ANALYZE is intentionally skipped due to external location constraints
- Data is read by path for downstream Gold processing
