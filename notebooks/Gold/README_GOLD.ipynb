# Gold Layer – Analytics-Ready Data

## Purpose
The Gold layer contains curated, analytics-ready fact and dimension tables
modeled using a star schema. These datasets are optimized for BI tools,
ad-hoc analytics, and downstream reporting.

## Design Principles
- Star schema with conformed dimensions
- Surrogate keys only in fact tables
- Strict grain enforcement per table
- Incremental, partition-aware Delta writes
- Idempotent and backfill-safe pipelines

## Fact Tables

### Fact_Airline_Tickets
- Grain: One row per airline market ticket
- Keys: Date, Origin Airport, Destination Airport, Carrier
- Measures: Passengers, fares, miles flown

### Fact_Route_Quarterly_Aggregation
- Grain: One row per origin–destination–carrier–quarter
- Measures:
  - Total passengers
  - Average market fare
  - Total market miles flown
  - Flight count
- Partitioning: source_year, source_quarter

## Dimension Tables
- Dim_Date
- Dim_Airport
- Dim_Carrier

## Operational Notes
- Late-arriving data is handled via partition-level reprocessing
- All Gold tables are derived strictly from Silver
- No raw or semi-processed data exists in Gold
